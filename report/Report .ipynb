# Отчет

В данном отчете мы проанализируем алгоритм, описанный в статье Baluja S., Rowley H. (2006) Boosting Sex Identification Performance. Данный метод идентифицирует пол человека по его фотографии. Результаты для этого отчета были получены с помощью программы, которая реализует этот алгоритм и написана на C++. 

Рассмотрим зависимость качества классификации от количества используемых слабых классификаторов. Построим графики соответствующих зависимостей:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_results(file_name):
    results = []
    with open(file_name, 'r') as f:
        for index, line in enumerate(f):
            if index % 3 == 2:
                results.append(np.float(line.split()[-1]))
    results = np.array(results)
    return results

directory = 'results/'
result_files = ['results_16_100_0.01.txt', 'results_16_100_0.1.txt']
plt.figure(figsize=(16, 12))
for index, result_file in enumerate(result_files):
    results = read_results(directory + result_file)
    plt.plot(np.arange(1, len(results[:100])+1), results[:100], label='{}% per iteration'.format(10**index))
plt.title('Accuracy vs. Number of weak classifiers')
plt.xlabel('Number of weak classifiers')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

Данные результаты представлены для размера патча $16\times 16$. По графику видно, что при уменьшении процента используемых классификаторов на каждой итерации, качество почти не понижается, и при увлечении числа используемых слабых классификаторов это понижение качества совсем исчезает. Рассмотрим, как ведет себя этот график при увеличении числа слабых классификаторов для случая, когда мы на каждой итерации используем один процент классификаторов. 

```python
result_files = ['results_16_1000_0.01.txt']
plt.figure(figsize=(16, 12))
for index, result_file in enumerate(result_files):
    results = read_results(directory + result_file)
    plt.plot(np.arange(1, len(results)+1), results, label='{}% per iteration'.format(10**index))
plt.title('Accuracy vs. Number of weak classifiers')
plt.xlabel('Number of weak classifiers')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

По этому графику виден общий тренд изменения качества при увеличении числа классификаторов. На данном эксперименте максимальное качество, которое было достигнуто, равно 0.777. 

Теперь рассмотрим результат аналогичного эксперимента для случая размера патча $20\times 20$. 

```python
result_files = ['results_20_100_0.01.txt']
plt.figure(figsize=(16, 12))
for index, result_file in enumerate(result_files):
    results = read_results(directory + result_file)
    plt.plot(np.arange(1, len(results)+1), results, label='{}% per iteration'.format(10**index))
plt.title('Accuracy vs. Number of weak classifiers')
plt.xlabel('Number of weak classifiers')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

Видим аналогичный возрастающий тренд. Максимальное достигнутое качество равно 0.782.

Проделаем аналогичное для размера патча $24\times24$.

```python
result_files = ['results_24_100_0.01.txt']
plt.figure(figsize=(16, 12))
for index, result_file in enumerate(result_files):
    results = read_results(directory + result_file)
    plt.plot(np.arange(1, len(results)+1), results, label='{}% per iteration'.format(10**index))
plt.title('Accuracy vs. Number of weak classifiers')
plt.xlabel('Number of weak classifiers')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

Получаем аналогичный результат с максимальным качеством, равным 0.804.

Сравним результаты каждого патча для случая, когда мы на каждой итерации используем один процент классификаторов:

```python
result_files = ['results_24_100_0.01.txt', 'results_20_100_0.01.txt', 'results_16_100_0.01.txt']
patch_sizes = [24, 20, 16]
plt.figure(figsize=(16, 12))
for index, result_file in enumerate(result_files):
    results = read_results(directory + result_file)
    plt.plot(np.arange(1, len(results)+1), results, label='{0}x{0}'.format(patch_sizes[index]))
plt.title('Accuracy vs. Number of weak classifiers')
plt.xlabel('Number of weak classifiers')
plt.ylabel('Accuracy')
plt.legend(loc='best')
```

По этому графику мы видим, что при увеличении размера патча качество улучшается. Это можно попытаться объяснить тем, что при этом увеличивается общее число классификаторов, из которых мы отбираем самых лучших в качестве слабых классификаторов. Но нужно учитывать тот факт, что при увеличении разрешения картинки сильно снижается скорость работы алгоритма.
